## Creating NumPY Arrays

-Creating arrays
-Creating Zeroes Matrix/Vector
-Creating All ones Matrix
-Creating Identity Matrix
-Creating Linearly Spaces Vector

```python
import numpy as np
```

```python
# Creating a 2D NumPy array
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr1
```

```python
# Creating a 2x3 array of zeroes
arr_zeroes = np.zeros((2,3))
arr_zeroes
```

```python
# Creating a 4x4 array of ones
arr_ones = np.ones((4,4))
arr_ones
```

```python
# Creating a 5x5 identity matrix
arr_identify = np.identity(5)
arr_identify
```

```python
#Creating an array with a range of values
arr2 = np.arange(10)
arr2
```

```python
#creating an array with a specific start, stop and step values

arr3 = np.arange(5, 15, 2)
arr3
```

```python
# Creating an array with linearly spaced values
arr4 = np.linspace(10, 100, 10)
arr4
```
