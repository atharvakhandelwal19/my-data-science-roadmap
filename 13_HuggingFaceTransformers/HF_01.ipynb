```python
from huggingface_hub import login
import os
from dotenv import load_dotenv
load_dotenv()

my_token = os.getenv("HF")

login(token=my_token)
```

```python
from transformers import pipeline
# import warnings

# warnings.filterwarnings("ignore")

classifier = pipeline("sentiment-analysis")

classifier(["I am feeeling Super tired today!", "I am gonna eat a Cake"])
```

```python
classifier= pipeline("zero-shot-classification")
classifier("This is a tutorial of Transformer Library", candidate_labels=["education", "politics", "business"])
```

## Generator

```python
generator = pipeline("text-generation", model="distilgpt2")
generator("Hello, My name is Atharv and I am 21 years old, Currently I am studying", max_length=40, num_return_sequences=3)
```

## UnMasker

```python
unmasker = pipeline("fill-mask")

unmasker("My dream is to go to <mask> for my masters. It's the best country and really beautiful", top_k=2)
```
