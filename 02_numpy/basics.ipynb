## NumPy Lists vs Python List Efficiency

```python
lista = list(range(100))
arraya = np.arange(100)
```

```python
import sys
```

```python
print(sys.getsizeof(lista)) 
print(sum(sys.getsizeof(x) for x in lista)) 

#For Arrays
print('Arrays')
print(arraya.nbytes) 
print(sys.getsizeof(arraya))
```

```python
import time
```

```python
x = list(range(100000000))
y = list(range(100000000, 200000000))

start = time.time()
c = [x[i] + y[i] for i in range(100000000)]
print("python math:", time.time() - start)
```

```python
a = np.arange(100000000)
b = np.arange(100000000, 200000000)

start = time.time()
c = a + b
print("numpy only add:", time.time() - start)
```

```python
arr11 = np.random.randint(low=1, high=100, size=20).reshape(4,5)
arr11
```

```python
# Indexing using boolean arrays
arr11[arr11>50]
```

```python
arr11[(arr11>50) & ~(arr11%2==0)]
```

```python
arr11[(arr11>50) & ~(arr11%2==0)] = 0
```
