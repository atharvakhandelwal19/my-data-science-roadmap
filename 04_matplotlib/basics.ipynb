## Using Data from CSV Files

```python
import pandas as pd
import csv
from collections import Counter

with open ('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    langauge_counter = Counter()    
    for row in csv_reader:
        langauge_counter.update(row['LanguagesWorkedWith'].split(';'))

langauges = []
popularity = []

for item in langauge_counter.most_common():
    langauges.append(item[0])
    popularity.append(item[1])

langauges.reverse()
popularity.reverse()
# df = pd.read_csv('data.csv', index_col='Responder_id')
# df['LanguagesWorkedWith'].str.split(';')
```

```python
plt.barh(langauges, popularity, label='Programming Languages')
plt.title('Most Popular Programming Languages')
plt.xlabel('Number of People Who Use')
```

## Using Pandas

```python
import pandas as pd
from collections import Counter

data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']


langauge_counter = Counter()    
for reponse in lang_responses:
    langauge_counter.update(reponse.split(';'))

langauges = []
popularity = []

for item in langauge_counter.most_common():
    langauges.append(item[0])
    popularity.append(item[1])

langauges.reverse()
popularity.reverse()
# df = pd.read_csv('data.csv', index_col='Responder_id')
# df['LanguagesWorkedWith'].str.split(';')
```

```python
plt.barh(langauges, popularity, label='Programming Languages')
plt.title('Most Popular Programming Languages')
plt.xlabel('Number of People Who Use')
```
