```python
import torch
import torch.nn as nn
import math
```

```python
pe = torch.zeros(3,512)
pe
```

```python
position= torch.arange(0, 3, dtype=torch.float).unsqueeze(1)
position
```

```python
div_term = torch.exp(torch.arange(0,512,2).float() * (-math.log(10000)/512))
div_term
```

```python
pe[:, 0::2] = torch.sin(position*div_term)
pe[:, 1::2] = torch.cos(position*div_term)

```
