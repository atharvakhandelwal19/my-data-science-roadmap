## Relational Plot

```python
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

## Scatter plot

```python
#Axis Level
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
#Figure Level plot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

## Line Plot

```python
gap = px.data.gapminder()

indiad = gap[gap['country'] == 'India']
indiad
```

```python
#axes level
sns.lineplot(data=indiad, x='year', y='pop')
```

```python
#relplot -> figure level

sns.relplot(data=indiad, x='year', y='gdpPercap', kind='line')
```

```python
##applying style
```

```python
data = gap[gap['country'].isin(['India', 'Brazil', 'Netherlands'])]
data.sample(10)
```

```python
sns.relplot(data=data, x='year', y='gdpPercap', hue='country', style='continent', kind='line')
```

## Matrix Plot

- Heat Map
- Cluster Map

```python
gap.head()
```

```python
temp_df = gap.pivot(index='country', columns='year', values='lifeExp')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
europe_df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
plt.figure(figsize=(10,10))
sns.heatmap(europe_df, annot=True, linewidths=0.5, cmap='viridis')
```

## MultiGrid Plots

- Facet Grid
- Pair Grid
- Joint Grid

```python
#Facet Grid

g = sns.FacetGrid(data=tips, col='day', row='time', hue='smoker')
g.map(sns.violinplot, 'sex', 'total_bill')
g.add_legend
```

## Pair Plot & Pair Grid

```python
iris.sample(10)
```

```python
sns.pairplot(iris, hue='species')
```

```python
g = sns.PairGrid(data = iris, hue="species")
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris, hue='species' ,vars=['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_offdiag(sns.kdeplot)
```

## Joint Plot and Joint Grid

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

```python
g = sns.JointGrid(data=tips, x= 'total_bill', y= 'tip')
g.plot(sns.scatterplot, sns.histplot)
```
