## Relational Plot

```python
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset("tips")
```

## Scatter plot

```python
#Axis Level
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
#Figure Level plot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

## Line Plot

```python
gap = px.data.gapminder()

indiad = gap[gap['country'] == 'India']
indiad
```

```python
#axes level
sns.lineplot(data=indiad, x='year', y='pop')
```

```python
#relplot -> figure level

sns.relplot(data=indiad, x='year', y='gdpPercap', kind='line')
```

```python
##applying style
```

```python
data = gap[gap['country'].isin(['India', 'Brazil', 'Netherlands'])]
data.sample(10)
```

```python
sns.relplot(data=data, x='year', y='gdpPercap', hue='country', style='continent', kind='line')
```

```python
##Getting Multiple Graphs using rel
```

```python
tips.head()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', row='smoker', col='sex', hue='sex', kind='scatter')
```

## Distribution Plot
. Histogram
. Kdeplot
. Rugplot

```python
#figure level -> displot
#axes level -> hist pde rug
```

```python
# sns.histplot(data=tips, x='total_bill', bins=5)
sns.histplot(data=tips, x='day')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic.sample(5)
```

```python
sns.displot(data=titanic, x='age', kind='hist', hue='sex', element='step')
```

```python
sns.displot(data=titanic, x='age', kind='hist', col='sex', element='step')
```

## KDE Plot

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```

## Rug Plot

```python
sns.kdeplot(data=tips, x='total_bill', fill=True)
sns.rugplot(data=tips, x='total_bill')
```

```python
#bivariate
sns.histplot(data=tips, x='total_bill', y='tip')
```

## Matrix Plot

- Heat Map
- Cluster Map

```python
gap.head()
```

```python
temp_df = gap.pivot(index='country', columns='year', values='lifeExp')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
europe_df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
plt.figure(figsize=(10,10))
sns.heatmap(europe_df, annot=True, linewidths=0.5, cmap='viridis')
```
