## KNN Classifier

```python
import pandas as pd
from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('car_evaluation.csv', names = ['buying', 'maint', 'doors', 'persons', 'lug_boots', 'safety', 'class'])
data.head()
```

```python
X = data[['buying', 'maint', 'safety']].values
y = data['class']
```

```python
print(X,y)
```

### Converting Classifiers into numbers

```python
# Conversion for X

Le = LabelEncoder()
for i in range(X.shape[1]):
    X[:, i] = Le.fit_transform(X[:, i])
```

```python
# Conversion for Y
y.unique()
```

```python
label_mapping = {'unacc': 0, 'acc': 1, 'good': 2, 'vgood': 3}

y = data['class'].map(label_mapping)
y.unique()
```

## K-neighbour Classifying

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=3, weights='distance')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

knn.fit(X_train,y_train)

predictions = knn.predict(X_test)
accuracy = metrics.accuracy_score(y_test, predictions)

print("predictions: ", predictions)
print("accuracy: ", accuracy)
```

## Support Vector Machine

```python
from sklearn import svm 
from sklearn.metrics import accuracy_score

iris= load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# print(X_train.shape)
# print(X_test.shape)
# print(y_train.shape)
# print(y_test.shape)

model = svm.SVC()
model.fit(X_train, y_train)

print(model)
predictions = model.predict(X_test)
acc = accuracy_score(y_test, predictions)

print(predictions)
print(acc)
```

## K-Means

```python
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
bc = load_breast_cancer()
```

```python
X = scale(bc.data)
y = bc.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
model = KMeans(n_clusters=2, random_state=0)
model.fit(X_train)
```

```python
predictions = model.predict(X_test)
labels = model.labels_
print('labels:', labels)
print('predictions:', predictions)
print('acc:', accuracy_score(y_test, predictions))
print('actual:', y_test)
```
